```python
import requests
from bs4 import BeautifulSoup as bs
import tkinter as tk

root = tk.Tk()
HEADERS = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; ) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36",
    "accept" : "*/*"
        }
root.geometry('800x600')

url = "https://www.olx.ua/zhivotnye/sobaki/lvov/"

def get_html(url, params = None):
    gethtml = requests.get(url, headers = HEADERS, params = params)
    return gethtml

def get_content(html):
    s = bs(html.text, "html.parser")
    return s


all_info = []
def parser():
    html = get_html(url)
    if html.status_code == 200:
        soup = get_content(html)
        items = soup.find_all("tr", class_ = "wrap")
        for item in items:
            priceTest = item.find("td", class_ = "td-price")
            if priceTest:
                price = item.find("p", class_ = "price").get_text(strip = True)
            else:
                price = "Ціну уточнюйте"
            all_info.append({"title" : item.find("a", class_ = "marginright5").get_text(strip = True),
                             "price" : price,
                             "city" : item.find("small", class_ = "breadcrumb").get_text(strip = True)
                            })
            prT = item.find("a", class_ = "marginright5").get_text(strip = True)
            prP = price
            prC = item.find("span").get_text(strip = True)
            text.insert(1.0, prT + prP + prC + "\n")
    else:
        print("Error")
        
        
parse = tk.Button(text = "Pasre", command = parser)
parse.pack(side = "top")
clean = tk.Button(text = "Clean",command = lambda : text.delete(1.0, "end"))
clean.pack(side = "top")
text = tk.Text()
scrollbar = tk.Scrollbar(command = text.yview )
scrollbar.pack( side = "right", fill = "y" )
text.config(yscrollcommand = scrollbar.set)
text.pack(side = "left", fill = "both")



root.mainloop()
```
